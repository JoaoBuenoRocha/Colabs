## Info geral:

**Iris L.**  um gnero de plantas com flor, muito apreciado pelas suas diversas espcies, que ostentam flores de cores muito vivas. So, vulgarmente, designadas como lrios, embora tal termo se aplique com mais propriedade a outro tipo de flor.  uma flor muito frequente em jardins. O termo ris  compartilhado, contudo, com outros gneros botnicos relacionados, da famlia Iridaceae. O termo pode ainda aplicar-se a uma subdiviso neste gnero.



## Podemos considerar os seguintes subgneros:

> **Iris**: com caule subterrneo (rizoma).

> **Limniris**: tambm com rizomas.

> **Xiphium**, por vezes considerado como um gnero  parte (Xiphion), que constitui o grupo principal de ris bulbosos.

> **Nepalensis**, por vezes considerado como o gnero,  parte, Junopsis; tambm bulboso.

> **Scorpiris**, por vezes considerado como o gnero Juno; tambm bulboso.

> **Hermodactyloides**, por vezes considerado como o gnero Iridodictyum, incluindo a pequena espcie Iris reticulata e outras semelhantes.


## Classificao cientfica:




> **Reino**:	Plantae

> **Diviso**:	Magnoliophyta

> **Classe**:	Liliopsida

> **Ordem**:	Asparagales

> **Famlia**:	Iridaceae

> **Gnero**:	Iris

```python
iris = df
```

```python
X = iris[["SepalWidthCm", "PetalWidthCm"]]
y = pd.get_dummies(iris.Species).values
y_classes = pd.get_dummies(iris.Species).columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                   test_size=0.3, 
                                                   random_state=seed)
```

```python
epochs = 200
batch = 64
model = build_model(input_shape, output_shape)
model.compile(loss = "categorical_crossentropy",
             optimizer = keras.optimizers.Adam(learning_rate=0.01),
             metrics = ["accuracy"])
history = model.fit(X_train, y_train, 
                   validation_data=(X_test, y_test),
                   verbose=True,
                   epochs= epochs, batch_size=batch)
```

```python
scores = model.evaluate(X_test, y_test, verbose=1)
print(f'Test accuracy: {scores[1]}')
```

```python
def plot_loss(history):
   plt.style.use("ggplot")
   plt.figure(figsize=(8, 4))
   plt.plot(history.history['loss'])
   plt.plot(history.history['val_loss'])
   plt.title("Model's training loss")
   plt.xlabel("Epoch #")
   plt.ylabel("Loss")
   plt.legend(['Train', 'Test'], loc='upper left')
   plt.show()


def plot_accuracy(history):
   plt.style.use("ggplot")
   plt.figure(figsize=(8, 4))
   plt.plot(history.history['accuracy'])
   plt.plot(history.history['val_accuracy'])
   plt.title("Model's training accuracy")
   plt.xlabel("Epoch #")
   plt.ylabel("Accuracy")
   plt.legend(['Train', 'Test'], loc='upper left')
   plt.show()
```

```python
plot_loss(history)
plot_accuracy(history)
```
